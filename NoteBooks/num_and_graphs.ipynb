```python
import pandas as pd
import numpy as np
df = pd.read_csv("data_cleaned.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Sales'], bins=50)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Profit'], bins=50, color="green")
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Discount", y="Profit")
plt.title("Discount vs Profit")
plt.show()
```

```python
top_cats = df.groupby("Product Category")['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cats.index, y=top_cats.values)
plt.title("Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
sub_profit = df.groupby("Product Sub-Category")['Profit'].sum().sort_values()

plt.figure(figsize=(12,5))
sns.barplot(x=sub_profit.index, y=sub_profit.values)
plt.title("Profit by Sub Category")
plt.xticks(rotation=45)
plt.show()
```

```python
top_states = df.groupby("Customer Segment")['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
sns.barplot(x=top_states.index, y=top_states.values)
plt.title("Sales By customer Segments")
plt.xticks(rotation=45)
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Customer Age'], bins=30, kde=True)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sub_profit = df.groupby("Product Sub-Category")['Profit'].sum().sort_values()

plt.figure(figsize=(12,5))
sns.barplot(x=sub_profit.index, y=sub_profit.values)
plt.title("Profit by Sub Category")
plt.xticks(rotation=45)
plt.show()

```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_order_quantity = df['Order Quantity'].sum()
average_discount = df['Discount'].mean()
print("total_sales, total_profit, total_order_quantity, average_discount")
print(total_sales, total_profit, total_order_quantity, average_discount)
```

```python
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
print("Profit_Margin")
df['Profit_Margin']
```

```python
category_analysis = df.groupby('Product Category')[['Sales','Profit','Discount','Order Quantity']].agg(['sum','mean','median'])
print("\n----- CATEGORY LEVEL ANALYSIS -----")
print(category_analysis)
```

```python
subcategory_analysis = df.groupby('Product Sub-Category')[['Sales','Profit','Order Quantity']].sum()
print("\n----- SUB-CATEGORY ANALYSIS -----")
print(subcategory_analysis.sort_values("Sales", ascending=False))
```

```python
region_analysis = df.groupby('Region')[['Sales','Profit']].sum()
print("\n----- REGION ANALYSIS -----")
print(region_analysis)
```

```python
region_analysis = df.groupby('State')[['Sales','Profit']].sum()
print("\n----- REGION ANALYSIS -----")
print(region_analysis)
```

```python
city_analysis = df.groupby('City')[['Sales','Profit']].sum()
print("\n----- CITY ANALYSIS -----")
print(city_analysis.sort_values("Sales", ascending=False))
```

```python
print("\n----- CUSTOMER AGE SUMMARY -----")
print(df['Customer Age'].describe())
```

```python
segment_summary = df.groupby('Customer Segment')[['Sales','Profit']].sum()
print("\n----- SEGMENT SUMMARY -----")
print(segment_summary)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

yearly_sales = df.groupby('Year')['Sales'].sum()
monthly_sales = df.groupby('Month')['Sales'].sum()

print("\n----- YEARLY SALES -----")
print(yearly_sales)

print("\n----- MONTHLY SALES -----")
print(monthly_sales)
```

```python
discount_effect = df.groupby(pd.cut(df['Discount'], bins=5))['Sales'].mean()
print("\n----- DISCOUNT EFFECT ON SALES -----")
print(discount_effect)
```

```python
corr = df[['Sales','Profit','Discount','Order Quantity','Unit Price','Shipping Cost']].corr()
print("\n----- CORRELATION MATRIX -----")
print(corr)
```

```python
def detect_iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper

for col in ['Sales','Profit','Order Quantity','Unit Price']:
    low, high = detect_iqr_outliers(df[col])
    outliers = df[(df[col] < low) | (df[col] > high)]
    print(f"\n----- OUTLIERS IN {col} -----")
    print(outliers[[col, 'Product Name', 'Product Category']])
```

```python
loss_products = df[df['Profit'] < 0][['Product Name','Product Category','Sales','Profit']]
print("\n----- LOSS MAKING PRODUCTS -----")
print(loss_products.sort_values("Profit"))
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("\n----- TOP 10 BESTSELLING PRODUCTS -----")
print(top_products)
```

```python
print("\n----- SHIPPING COST SUMMARY -----")
print(df['Shipping Cost'].describe())

ship_mode_analysis = df.groupby('Ship Mode')['Shipping Cost'].mean()
print("\nAVERAGE SHIPPING COST BY MODE:")
print(ship_mode_analysis)
```

```python
priority_analysis = df.groupby('Order Priority')[['Sales','Profit','Order Quantity']].sum()
print("\n----- PRIORITY ANALYSIS -----")
print(priority_analysis)
```

```python
container_analysis = df.groupby('Product Container')[['Sales','Profit']].sum()
print("\n----- CONTAINER ANALYSIS -----")
print(container_analysis)
```
